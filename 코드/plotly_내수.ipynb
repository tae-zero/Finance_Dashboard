```python
import pandas as pd

# 데이터 로드
df = pd.read_csv(r'C:\Users\bitcamp\OneDrive\Desktop\pythontraning\NICE_내수수출_코스피.csv')

# 피벗 테이블 생성
pivot_df = pd.pivot_table(
    df,
    index=['종목명', '사업부문', '매출품목명'],
    columns='구분',
    values=['2022_12 매출액', '2023_12 매출액', '2024_12 매출액'],
    aggfunc='sum',
    fill_value=0  # 결측값은 0으로
)

# 보기 쉽게 컬럼 정리
pivot_df.columns = ['_'.join([str(i) for i in col]).strip() for col in pivot_df.columns.values]
pivot_df = pivot_df.reset_index()

# 결과 확인
pivot_df.head()
```
